```python
#Q5
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\TUSHAR UPADHYAY\\Desktop\\Python\\Querise solution\\superstore.csv")
df['Discount'] = df['Discount'].str.rstrip('%').astype('float') / 100.0
average_discount = df.groupby('Region')['Discount'].mean().sort_values(ascending=False)
```

```python
plt.bar(average_discount.index, average_discount)
plt.xlabel('Region')
plt.ylabel('Average Discount')
plt.title('Average Discount across Regions')
plt.show()
highest_average_discount_region = average_discount.idxmax()
print("Region with the highest average discount:", highest_average_discount_region)
```

```python
#Analysis 
#Avergae Profit vs Ship mode
average_profit = df.groupby('Ship Mode')['Profit'].mean().sort_values()
plt.bar(average_profit.index, average_profit)
plt.xlabel('Ship Mode')
plt.ylabel('Average Profit')
plt.title('Average Profit by Ship Mode')
plt.xticks(rotation=45)
plt.show()
max_profit_shipmode = average_profit.idxmax()

print("Ship Mode with Maximum Profit:", max_profit_shipmode)
```

```python
#Average Profit vs Segment
average_profit = df.groupby('Segment')['Profit'].mean().sort_values()
plt.bar(average_profit.index, average_profit)
plt.xlabel('Segment')
plt.ylabel('Average Profit')
plt.title('Average Profit by Segment')
plt.xticks(rotation=45)
plt.show()
highest_profit_segment = average_profit.idxmax()
print("Segment with Highest Average Profit:", highest_profit_segment)

```

```python
#ship mode vs segment
ship_segment_counts = df.groupby(['Ship Mode', 'Segment']).size().unstack()
ship_segment_counts.plot(kind='bar', stacked=True)
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.title('Ship Mode vs Segment')
plt.legend(title='Segment')
plt.xticks(rotation=45)
plt.show()
```

```python
#Average discount vs Region 
average_discount = df.groupby('Region')['Discount'].mean().sort_values()

plt.bar(average_discount.index, average_discount)
plt.xlabel('Region')
plt.ylabel('Average Discount')
plt.title('Average Discount by Region')
plt.xticks(rotation=45)
plt.show()

highest_discount_region = average_discount.idxmax()
print("Region with Highest Average Discount:", highest_discount_region)
```

```python
#average profit vs region 
average_profit = df.groupby('Region')['Profit'].mean().sort_values()

plt.bar(average_profit.index, average_profit)
plt.xlabel('Region')
plt.ylabel('Average Profit')
plt.title('Average Profit by Region')
plt.xticks(rotation=45)
plt.show()
highest_profit_region = average_profit.idxmax()
print("Region with Highest Average Profit:", highest_profit_region)
```

```python
# Region vs ship mode
ship_mode_counts = df.groupby(['Region', 'Ship Mode']).size().unstack()

ship_mode_counts.plot(kind='bar', stacked=True)
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Region vs Ship Mode')
plt.legend(title='Ship Mode')
plt.xticks(rotation=45)
plt.show()
```

```python
#region vs segment 
segment_counts = df.groupby(['Region', 'Segment']).size().unstack()

segment_counts.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Region vs Segment')
plt.legend(title='Segment')
plt.xticks(rotation=45)
plt.show()
```
